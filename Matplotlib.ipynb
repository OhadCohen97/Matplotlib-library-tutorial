**Visualization with Matplotlib**

```python
import numpy as np
```

Importing Matplotlib library:

```python
import matplotlib.pyplot as plt
```

Lets write some simple date(global warming) for visualize it.

```python
year = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
temp_index = [0.72,0.61,0.65,0.68,0.75,0.90,1.02,0.93,0.85,0.99,1.02]
```

when using 'plot' method,we need to define x and y.

x=year, y=temp_index

```python
plt.plot(year,temp_index)
```

We should add some information to the plotted graph.

```python
plt.plot(year,temp_index)
plt.title('Global Warming',{'fontsize':15,'horizontalalignment':'center'})
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show() #help you for render the graph
```

Another simple date (customers using electricity):

```python
months=['Jan', 'Feb', 'March', 'April','May', 'June','July','Aug','Sep','Oct','Nov','Dec']
customer_1=[12,13,9,8,7,8,8,7,6,5,8,10]
customer_2=[14,16,11,7,6,6,7,6,5,8,9,12]
```

```python
plt.plot(months,customer_1)
plt.plot(months,customer_2)
plt.show()
```

As you can see, the plot is not informative and we can not know which customer belongs to each line in the graph.

Rewrite the code again with adding x and y labels, 'label','marker="o"'

```python
plt.plot(months,customer_1,label='customer_1',marker='o')
plt.plot(months,customer_2,label='customer_2',marker='^')
plt.xlabel('Months')
plt.ylabel('Electricity Consumption')
plt.title('Building Consumption')
plt.legend(loc='upper center') # showing the labels on the upper center of the graph.
plt.show()
```

**Scatter plot**

```python
plt.scatter(months,customer_1,color='red',label='customer_1')
plt.scatter(months,customer_2,color='blue',label='customer_2')
plt.xlabel('Months')
plt.ylabel('Electricity Consumption')
plt.title('Scatter plot of Building Consumption')
plt.grid()#squares will appear on the graph
plt.legend()
plt.show()
```

**Histogram plot**

Showing the number of occurrences of each of the data.
It actually shows you a distribution of your consumption.
Use 'hist' method for plotting Histogram graph.

Just mention one vector as y.

```python
plt.hist(customer_1,bins=20,color='green')#bins refers to thickness of the green lines

plt.ylabel('Electricity Consumption')
plt.title('Histogram')
plt.show()
```

**Bar Chart plot**

```python
plt.bar(months,customer_1,width=0.8)
plt.xlabel('Months')
plt.ylabel('Electricity Consumption')
plt.show()
```

Plot two separated Histograms:

```python
bar_width = 0.4
month_n =np.arange(12) #defining array with values 1 to 12

plt.bar(month_n,customer_1,bar_width,label='customer_1',color='blue')
plt.bar(month_n+bar_width,customer_2,bar_width,label='customer_2',color='red')#month_n+bar_width will shift right the customer_2's lines.
plt.xlabel('Months')
plt.ylabel('Electricity Consumption')
plt.show()
```

As you can see, we need to change back the axis label to 'Months'.

I'm using the 'xticks' method to change the axis labels

```python
bar_width = 0.4
month_n =np.arange(12) #defining array with values 1 to 12

plt.bar(month_n,customer_1,bar_width,label='customer_1',color='blue')
plt.bar(month_n+bar_width,customer_2,bar_width,label='customer_2',color='red')#month_n+bar_width will shift right the customer_2's lines.
plt.xlabel('Months')
plt.xticks(month_n+bar_width,months)
plt.ylabel('Electricity Consumption')
plt.show()
```

**Box plot**

'notch=False' will plot simple form of rectangle box plot.

'vert=False' is vertical box plot,while it set to False, it will give us a horizontal box

```python
plt.boxplot(customer_1,notch=False,vert=False)
```

A box and whisker plot—also called a box plot—displays the five-number summary of a set of data.

The five-number summary is the minimum, first quartile, median, third quartile, and maximum.

In a box plot, we draw a box from the first quartile to the third quartile.
A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.

```python
plt.boxplot([customer_1,customer_2],patch_artist=True,
            boxprops=dict(facecolor='red',color='red'),
            whiskerprops=dict(color='green'),
            capprops=dict(color='blue'),
            medianprops=dict(color='yellow'))
plt.show()
```
